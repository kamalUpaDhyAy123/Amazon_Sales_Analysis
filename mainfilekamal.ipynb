```python
#Calculate the attrition rate
data['Attrition'] = data['Attrition'].map({'Yes': 1, 'No': 0})
attrition_rate = data['Attrition'].mean() * 100
print(f"Attrition Rate: {attrition_rate:.2f}%")
```

```python
# Analyze the factors: Age, Years at Company, Income
# Calculate mean values for employees who left and those who stayed
attrition_factors = data.groupby('Attrition').agg({
    'Age': 'mean',
    'YearsAtCompany': 'mean',
    'MonthlyIncome': 'mean'
}).reset_index()

print(attrition_factors)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plots
sns.set(style="whitegrid")

# Box plots for each factor
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.boxplot(x='Attrition', y='Age', data=data, ax=axes[0])
axes[0].set_title('Age vs Attrition')

sns.boxplot(x='Attrition', y='YearsAtCompany', data=data, ax=axes[1])
axes[1].set_title('Years at Company vs Attrition')

sns.boxplot(x='Attrition', y='MonthlyIncome', data=data, ax=axes[2])
axes[2].set_title('Income vs Attrition')

plt.tight_layout()
plt.show()
```

```python
from scipy import stats

# Example: t-test for Age
left_age = data[data['Attrition'] == True]['Age']
stayed_age = data[data['Attrition'] == False]['Age']
t_stat, p_value = stats.ttest_ind(left_age, stayed_age)
print(f"t-statistic: {t_stat}, p-value: {p_value}")

# Logistic Regression to predict attrition
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Prepare data for logistic regression
X = data[['Age', 'YearsAtCompany', 'MonthlyIncome']]
y = data['Attrition'].astype(int)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print(classification_report(y_test, y_pred))


```
